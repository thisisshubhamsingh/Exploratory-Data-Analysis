## IRIS dataset EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

### Qualitative features and Quantitative features

```python
list(df.select_dtypes(exclude = object).columns.values)
```

```python
list(df.select_dtypes(include = object).columns.values)
```

* Qualitative feature:  1
* Quantitative feature: 4

#### Unique species 

```python
list(df['species'].unique())
```

```python
df.isnull().sum()
```

### Frequency Table

```python
freq_table = df.groupby(df['species']).size()
print(freq_table)
```

### Sorting numerical features 

```python
df.sort_values(by = ['sepal_length','sepal_width','petal_length' ,'petal_width'] , axis = 0 , ascending = True , inplace=True)

df.reset_index(drop=True , inplace=True)
```

```python
df.head()
```

### Descriptive Statistics

```python
df.describe()
```

#### Datatype check

```python
df.dtypes
```

### Correlation matrix

```python
sns.heatmap(data = df.iloc[:,0:4].corr() , annot=True )
```

So here we can see that the petal length and petal width is highly correlated

### Visualization of Correlation

```python
sns.pairplot(data = df)
```

petal length and petal width has positive high correlation

### Distribution of the data

```python
fig , ax = plt.subplots(2,2,figsize = (10,6))


sns.histplot(ax = ax[0,0],data = df['sepal_length'] , kde = True)
sns.histplot(ax = ax[0,1],data = df['sepal_width'] , kde = True)
sns.histplot(ax = ax[1,0],data = df['petal_length'] ,kde = True)
sns.histplot(ax = ax[1,1],data = df['petal_width'] ,kde = True)
```

Here we can clearly see that the sepal width feature has been following gaussian distribution.
